```python
final = [
    ["1", "2", "3"],
    ["4", "5", "6"],
    ["7", "8", None],
]
```

```python
def move_left(initial):

    temp_state = [
        ["0", "0", "0"],
        ["0", "0", "0"],
        ["0", "0", "0"],
    ]

    temp = 0

    for i in range(3):
        for j in range(3):
            temp_state[i][j] = initial[i][j]

    for i in range(0, 2):
        for j in range (0, 2):
            if initial[i][j] == None:
                print(i, j)
                if j >= 1:
                    temp = temp_state[i][j-1]
                    temp_state[i][j-1] = temp_state[i][j]
                    temp_state[i][j] = temp
                    print(temp_state)
                    break
                else:
                    print("Can't move left!")

    print(move_left(initial))
```

```python
def move_right(initial):

    temp_state = [
        ["0", "0", "0"],
        ["0", "0", "0"],
        ["0", "0", "0"],
    ]

    temp = 0

    for i in range(3):
        for j in range(3):
            temp_state[i][j] = initial[i][j]

    for i in range(0, 2):
        for j in range (0, 2):
            if initial[i][j] == None:
                if j <= 1:
                    temp = temp_state[i][j+1]
                    temp_state[i][j+1] = temp_state[i][j]
                    temp_state[i][j] = temp 
                    print(temp_state)
                    break
                else:
                    print("Can't move right!")

    print(move_right(initial))
```

```python
def move_up(initial):

    temp_state = [ 
        ["0", "0", "0"],
        ["0", "0", "0"],
        ["0", "0", "0"],
    ]

    temp = 0

    for i in range(3):
        for j in range(3):
            temp_state[i][j] = initial[i][j]

    for i in range(0, 2):
        for j in range (0, 2):
            if initial[i][j] == None:
                if i >= 1:
                    temp = temp_state[i-1][j]
                    temp_state[i-1][j] = temp_state[i][j]
                    temp_state[i][j] = temp
                    print(temp_state)
                    break
                else:
                    print("Can't move up!")

    print(move_up(initial))
```

```python
def move_down(initial):

    temp_state = [
        ["0", "0", "0"],
        ["0", "0", "0"],
        ["0", "0", "0"],
    ]

    temp = 0

    for i in range(3):
        for j in range(3):
            temp_state[i][j] = initial[i][j]

    for i in range(0, 2):
        for j in range (0, 2):
            if initial[i][j] == None:
                if i <= 1:
                    temp = temp_state[i+1][j]
                    temp_state[i+1][j] = temp_state[i][j]
                    temp_state[i+1][j] = temp
                    print(temp_state)
                    break
                else:
                    print("Can't move down!")

    print(move_down(initial))
```
